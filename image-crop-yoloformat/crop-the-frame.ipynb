```python
import cv2
import os
import numpy as np
from  matplotlib import pyplot as plt
from shutil import copyfile
%matplotlib inline
```

```python
def plot_image(image, title):
    plt.imshow(image)
    plt.title(title, size=20)
    plt.axis('off')
```

```python
basedir = "wzy/"
yolodir = basedir+"yolo/"
```

```python
def red_frame(hsv):
    low_hsv = np.array([156,43,46])
    high_hsv = np.array([180,255,255])
    mask = cv2.inRange(hsv, lowerb=low_hsv, upperb=high_hsv)
    return mask

def blue_frame(hsv):
    low_hsv = np.array([100,43,46])
    high_hsv = np.array([124,255,255])
    mask = cv2.inRange(hsv, lowerb=low_hsv, upperb=high_hsv)
    return mask

def bbox2yolo(size, box):
    dw = 1./size[0]
    dh = 1./size[1]
    x = (box[0] + box[2]/2.0)*dw
    y = (box[1] + box[3]/2.0)*dh
    w = box[2]*dw
    h = box[3]*dh
    return (x,y,w,h)

def find_contours_save_yolo(hsv, img_name):
    with open(yolodir+img_name+".txt",mode='w+',encoding='utf-8') as f:
        mask = red_frame(hsv)
        contours, hier = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        for c,i in zip(contours, range(0, len(contours))):  #遍历轮廓
            yolo = bbox2yolo(hsv.shape, cv2.boundingRect(c))
            yolostr = "0 " + " ".join(str(i) for i in yolo)+"\n"
            f.writelines(yolostr)
        mask = blue_frame(hsv)
        contours, hier = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        for c,i in zip(contours, range(0, len(contours))):  #遍历轮廓
            yolo = bbox2yolo(hsv.shape, cv2.boundingRect(c))
            yolostr = "1 " + " ".join(str(i) for i in yolo)+"\n"
            f.writelines(yolostr)
        
```

```python
# 多张图，多个框,多种颜色框
dir_list = [f for f in os.listdir(basedir) if not f.startswith(('.', 'yolo'))]
for dir in dir_list:
    print(dir)
    img_list = [f for f in os.listdir(basedir+dir+"/bbox/") if not f.startswith('.')]
    for img in img_list:
        print("Now processing image: "+ dir + "/" + img)
        img_name = os.path.splitext(img)[0]
        dir_img_name = dir+"-"+img_name
        sample_img = cv2.imread(basedir+dir+"/bbox/"+img)
        origin_img_dir  = basedir+dir+"/png/"+img
        hsv = cv2.cvtColor(sample_img, cv2.COLOR_BGR2HSV)
        find_contours_save_yolo(hsv, dir_img_name)
        copyfile(origin_img_dir, yolodir+dir+"-"+img)
print("All processing image done !!")
```
